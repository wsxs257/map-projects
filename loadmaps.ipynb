```python
import PIL
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
xdim = 16150 + 512
ydim = 11512 + 512
import matplotlib.image as mpimg
img = np.zeros((ydim, xdim, 3),dtype = 'uint8')
for horiz in range(0, xdim-512, 512):
    print('{} / {} complete.'.format(horiz, xdim))
    for vert in range(0, ydim-512, 512):
        filename = 'https://map-view.nls.uk/iiif/2/11448%2F114486787/' + str(horiz) + ',' + str(vert) + ',512,512/512,/0/default.jpg'
        imload= mpimg.imread(filename, 'jpg')
        imshape = imload.shape
        img[vert:vert+imshape[0], horiz:horiz+imshape[1], :] = imload 
plt.imshow(img)
```

```python
plt.imsave('coton1.png',img)
```
